```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import os
from pathlib import Path


import warnings
warnings.filterwarnings('ignore')
```

```python
def process_file(file_path):
    df = pd.read_csv(file_path)
    filtered_df = df[(df['RmTmpCspt'] < 85) & (df['RmTmpHpst'] > 65)]
    
    if filtered_df.empty:
        print(f"No matching data for {file_path}")
        return

    filtered_df['timestamp'] = pd.to_datetime(filtered_df['timestamp'])
    filtered_df['date'] = filtered_df['timestamp'].dt.date
    filtered_df=filtered_df[50000:300000]

    filtered_df['time_diff'] = filtered_df['timestamp'].diff()
    filtered_df['new_interval'] = (filtered_df['time_diff'] != pd.Timedelta(minutes=15)).cumsum()
    filtered_df['new_interval'] = filtered_df['new_interval'][0:20]
    
    for interval, group in filtered_df.groupby('new_interval'):
        first_date = group['timestamp'].iloc[0].strftime('%Y-%m-%d')
        print(f"File: {file_path}, Date: {first_date}")
        print(group)

        # Plotting
        fig, ax1 = plt.subplots(figsize=(10, 6))  # Create figure and axes
        ax1.plot(group['timestamp'], group['RmTmp'], color='black', label='RmTmp',marker='o')
        ax1.plot(group['timestamp'], group['RmTmpCspt'], color='red', label='Cooling Set Point (RmTmpCspt)',marker='o')
        ax1.plot(group['timestamp'], group['RmTmpHpst'], color='blue', label='Heating Set Point (RmTmpHpst)',marker='o')

        ax1.set_ylim(60, 80)  # Set y-axis limits for ax1
        ax1.set_xlabel("Timestamp")  # Add x-axis label
        ax1.set_ylabel("Temperature (°F)")  # Add y-axis label for ax1
        ax1.legend(loc='upper left')  # Add legend for ax1

        ax2 = ax1.twinx()  # Create a twin y-axis
        ax2.plot(group['timestamp'], group['HwVlvPos'], color='green', label='HwVlvPos',marker='o')
        ax2.set_ylabel("Valve Position")  # Add y-axis label for ax2
        ax2.legend(loc='upper right')  # Add legend for ax2

        plt.title(f"{file_path} - Date: {first_date}")  
        plt.grid()  
        plt.xticks(rotation=45)  
        plt.tight_layout() 
        plt.show()


# Traverse the files
base_path = Path("data/combined_milas_hall")
for x in [2, 3]:
    if x == 2:
        numbers = range(25, 34)
    elif x == 3:
        numbers = range(62, 71)
    
    for number in numbers:
        file_name = f"Flo2.3-A{x}-{number}.csv"
        file_path = base_path / file_name
        if file_path.exists():
            process_file(file_path)
        else:
            print(f"File not found: {file_path}")
```
